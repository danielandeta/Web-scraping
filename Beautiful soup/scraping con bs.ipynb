```python
from requests import get
url = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
response = get(url)
print(response.text[:500])
```

```python
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
contenedores_peliculas = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(contenedores_peliculas))
print(len(contenedores_peliculas))
```

```python

# Listas para guardar la data extraída
nombres = []
anios = []
imdb_ratings = []
metascores = []
votos = []
# Extracción de data de los contenedores de películas 
for contenedor in contenedores_peliculas:
# Si la película tiene Metascore, extraer: 
    if contenedor.find('div', class_ = 'ratings-metascore') is not None:
# nombre
        nombre = contenedor.h3.a.text
        nombres.append(nombre)
# año
        anio = contenedor.h3.find('span', class_ = 'lister-item-year').text
        anios.append(anio)
# IMDB rating
        imdb = float(contenedor.strong.text)
        imdb_ratings.append(imdb)
# metascore
        m_score = contenedor.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
# número de votoss
        voto = contenedor.find('span', attrs = {'name':'nv'})['data-value']
        votos.append(int(voto))
```

```python
import pandas as pd
test_df = pd.DataFrame({'película': nombres,
'años': anios,
'imdb': imdb_ratings,
'metascore': metascores,
'votos': votos
})
print(test_df.info())
test_df
```
